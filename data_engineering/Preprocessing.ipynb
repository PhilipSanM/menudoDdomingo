```python
import nltk
import spacy
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords

import re
import contractions
from bs4 import BeautifulSoup

#import cupy as cp
import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer

nltk.download('punkt')
nltk.download('stopwords')
```

```python
def process_and_save_text(input_path, output_path):
    with open(input_path, 'r', encoding='utf-8') as file:
        document = file.read()

    soup = BeautifulSoup(document, "html.parser")
    doc_html = soup.get_text()

    doc_low = doc_html.lower()
    text = doc_low

    sentences = sent_tokenize(text)

    sentences_space = [re.sub(r'\s+([^\w\s])', r'\1', sentence) for sentence in sentences]
    sentences_url = [re.sub(r'http\S+', '', sentence) for sentence in sentences_space]
    sentences_chars = [re.sub(r'[^a-záéíóúüñ\s]', '', sentence) for sentence in sentences_url]

    words = [word_tokenize(sentence) for sentence in sentences_chars]

    stop_words = set(stopwords.words('spanish'))

    filtered_sentences = [[word for word in sentence if word.lower() not in stop_words] for sentence in words]

    with open(output_path, 'w', encoding='utf-8') as output_file:
        for sentence in filtered_sentences:
            output_file.write(' '.join(sentence) + '\n')

    print(f"Resultado guardado en '{output_path}'")
```

```python
for root, rt in zip(roots, rts):
  process_and_save_text(root, rt)
```
